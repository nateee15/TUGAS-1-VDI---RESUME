#**MAKING DATA VISUALIZATION MORE EFFECIENT AND EFFECTIVE**

###**Apasih itu visuaisasi Data?**

Visualisasi data adalah suatu kegiatan atau proses yang dapat mengubah data abstrak jadi bentuk visual seperti panjang, posisi, bentuk, dan warna. Hal ini merupakan cara yang sangat efektif untuk menyampaikan cerita dari data kepada orang-orang yang lebih suka visual.

APASIH ITU.jpeg

Dewasa ini, hampir semua organisasi punya lebih banyak data daripada sebelumnya. Oleh karena itu, makin banyak organisasi yang menggunakan data dan analitik canggih untuk membantu pengambilan keputusan strategis dan operasional. Visualisasi data sendiri sangat cocok untuk memberikan gambaran umum yang jelas tentang data yang besar, dan membuat hasil analisis lebih mudah dipahami oleh ilmuwan data.



###**Bagaimana Perkembangan Visualisasi Data?**

Perkembangan visualisasi data telah mencapai kemajuan pesat berkat kontribusi berbagai komunitas, termasuk komunitas grafik komputer yang telah mengembangkan teknologi rendering canggih seperti D3 dan VegaLite. Komunitas visualisasi data telah mempermudah pengguna untuk membuat dan berinteraksi dengan visualisasi melalui alat seperti Tableau dan Microsoft Power BI, yang didukung oleh teknologi backend seperti Hyper DB dan Apache Spark. Komunitas basis data juga telah meningkatkan pengalaman pengguna dengan memungkinkan visualisasi real-time untuk data besar, termasuk jutaan hingga miliaran catatan. Visualisasi data kini telah menjadi bagian penting dalam berbagai aplikasi, seperti Excel, Google Sheets, dan Amazon Quicksight, yang digunakan secara luas di industri.


###**Pipeline visualisasi**

Pipeline visualisasi data terdiri dari beberapa langkah penting, mulai dari pengambilan data, persiapan, manipulasi, hingga rendering menjadi representasi visual. Dalam pengembangan visualisasi data, terdapat tiga arah utama yang berperan penting:
1. **Spesifikasi Visualisasi**: Menyediakan berbagai cara bagi pengguna untuk menentukan visualisasi yang diinginkan, seperti melalui bahasa pemetaan dan manipulasi data. Hal ini membantu pengguna memahami cara menghasilkan visualisasi data yang efektif.
2. **Pendekatan Efisien untuk Visualisasi Data**: Pentingnya proses yang efisien dan skalabel dalam membuat visualisasi, terutama untuk manipulasi data dan pemetaan, termasuk integrasi dengan sistem pemrosesan data seperti SQL dan HyperDB. Solusi pendekatan ini membantu dalam penanganan big data untuk memberikan respons real-time.
3. **Rekomendasi Visualisasi Data**: Mengingat kompleksitas dalam menentukan visualisasi yang tepat, beberapa sistem menawarkan rekomendasi otomatis yang membantu pengguna dalam memilih visualisasi yang paling sesuai dengan data dan cerita yang ingin disampaikan.

###**Spesifikasi Visualisasi Data**

Umumnya, bahasa visualisasi data terdiri dari tiga bagian utama: data, tanda visual (atau petunjuk visual), dan pemetaan antara keduanya.
1.	**Data:**
    * Records: Data yang akan divisualisasikan.
    * Transformasi: Operasi seperti grup, bin, filter, dan sort yang digunakan untuk mengubah data yang telah ditentukan.
2.	**Tanda Visual (atau Petunjuk Visual)**:
    * Tipe: Representasi visual dari data, seperti bar, garis, atau titik.
    * Ukuran: Lebar dan tinggi dari visualisasi.
    * Legenda: Informasi legenda yang menjelaskan elemen visual.
    * Lain-lain: Properti lain, seperti lebar dan warna bar.
3.	**Pemetaan**:
    * Menghubungkan data dengan tanda visual yang sesuai.
Operasi visual berbasis GUI biasanya diterjemahkan ke dalam bahasa visualisasi data ini.


###**Klasifikasi Bahasa Visualisasi Data**

Strategi umum untuk mengklasifikasikan bahasa visualisasi data didasarkan pada tingkat ekspresivitasnya. Semakin rendah tingkat bahasa, semakin ekspresif bahasa tersebut. Bahasa tingkat tinggi menyederhanakan beberapa detail tingkat rendah dengan menyediakan default yang masuk akal dan menambahkan lebih banyak batasan (misalnya, Excel menyediakan template untuk visualisasi yang didukung). Dimensi lain untuk memahami tingkat bahasa spesifikasi visualisasi adalah aksesibilitasnya; semakin tinggi tingkat bahasa, semakin mudah digunakan.

* **Bahasa Tingkat Rendah** : Bahasa tingkat rendah mengharuskan pengguna untuk menentukan semua elemen pemetaan. Contohnya termasuk Prefuse dan Flare, yang merupakan pustaka visualisasi berbasis Java. Bahasa ini mengenkapsulasi item visual sebagai kelas Java dengan banyak atribut visual, dan pengguna memetakan data ke atribut visual ini dengan menetapkan fungsi yang telah ditentukan. Protovis adalah toolkit grafis berbasis JavaScript yang menggunakan tanda grafis sederhana (bar, area, garis, dll.) dengan atribut visual yang ditentukan. D3 adalah pengembangan dari Protovis yang lebih efektif dalam menangani interaksi pengguna. Vega dan Reactive Vega mirip dengan Protovis dan D3, tetapi mereka menyediakan tata bahasa interaksi yang dapat dikomposisi secara deklaratif.

* **Bahasa Tingkat Tinggi** : Bahasa tingkat tinggi mengenkapsulasi detail konstruksi visualisasi, seperti fungsi pemetaan, serta beberapa properti untuk tanda visual seperti ukuran kanvas, legenda, dan properti lainnya. ggplot2 dibangun di atas karya Wilkinson pada tahun 2005 "The Grammar of Graphics"; ini adalah tata bahasa grafis berlapis yang tertanam dalam bahasa R. Vega-Lite adalah pengembangan lebih lanjut dari Vega dan Reactive Vega, mendukung desain interaksi yang dapat dikomposisi tetapi menyediakan tata bahasa yang lebih ringkas. Altair membuat Vega-Lite tersedia untuk komunitas Python. Echarts adalah pengembangan terbaru dalam bahasa visualisasi deklaratif yang dirancang untuk mendukung pembuatan visualisasi cepat bagi non-programmer. VizQL, yang berkembang dari sistem Polaris, adalah bahasa spesifikasi visualisasi dari Tableau. ZQL dari Zenvisage menggunakan bahasa struktur tabular di mana setiap baris dalam tabel adalah spesifikasi visualisasi.

### **Operasi Visual Berbasis GUI**

Dibandingkan dengan menggunakan bahasa visualisasi deklaratif untuk menentukan visualisasi, cara yang lebih ramah pengguna adalah mengikuti prinsip "manipulasi langsung," yang merupakan konsep yang banyak digunakan dalam interaksi manusia-komputer. Alat-alat GUI canggih seperti Tableau, Qlik, Excel, Google Sheets, Amazon Quicksight, Microsoft Power BI, dan lainnya mempermudah pengguna dalam menentukan visualisasi.



###**Visualisasi Data Interaktif**

Dasar dari visualisasi data interaktif adalah bahwa visualisasi data seringkali merupakan proses eksplorasi, di mana pengguna perlu terus menyempurnakan spesifikasi visualisasi yang sedang dieksplorasi hingga mereka mendapatkan visualisasi yang diinginkan. Polaris dan Tableau, misalnya, memungkinkan penyempurnaan kueri secara bertahap untuk menciptakan visualisasi multidimensi. Selain itu, alat seperti DeepEye dan Voyager memudahkan eksplorasi visualisasi dengan navigasi berfaktor.

* **Penyempurnaan Kueri Bertahap**: Polaris dan Tableau menyediakan template grafik untuk menunjukkan visualisasi multidimensi dalam struktur tabular. Struktur ini memungkinkan perbandingan dan analisis atribut yang berbeda secara mudah. Pengguna dapat secara bertahap menambahkan atribut ke baris, kolom, dan lapisan visualisasi, memilih jenis visualisasi yang sesuai, dan memetakan data ke properti visual untuk membangun visualisasi yang diinginkan langkah demi langkah.

* **Navigasi Berfaktor**: DeepEye mendukung navigasi berfaktor untuk membantu pengguna menjelajahi ruang desain visualisasi. Pengguna dapat mengetikkan kata kunci, dan DeepEye akan merekomendasikan visualisasi yang relevan. Voyager juga menawarkan eksplorasi ruang visualisasi dengan merekomendasikan visualisasi yang memiliki atribut serupa atau lebih banyak.

###**Spesifikasi yang Tidak Lengkap**

Visualisasi data tidak memiliki makna jika tidak dapat memberikan wawasan dari data tersebut. Namun, seringkali pengguna tidak mengetahui semua aspek dari data yang dimiliki, terutama ketika data tersebut besar dan sering diperbarui. Oleh karena itu, diperlukan dukungan untuk spesifikasi yang tidak lengkap.

Umumnya, untuk spesifikasi yang tidak lengkap, pengguna hanya memberikan beberapa "petunjuk", dan sistem visualisasi bertugas untuk menafsirkan input tersebut dengan berbagai cara.

1. **Berbasis Referensi**: Pengguna memberikan visualisasi referensi sebagai contoh, dan sistem kemudian menyarankan visualisasi lain yang serupa atau berbeda berdasarkan referensi tersebut. Contohnya adalah zenvisage yang mengembalikan visualisasi dengan tren serupa atau berbeda dari visualisasi referensi.

2. **Berbasis Kata Kunci**: Pengguna memberikan tujuan visualisasi melalui kata kunci, dan sistem merekomendasikan visualisasi yang sesuai. Misalnya, DeepEye menerima input kata kunci seperti "tampilkan grafik garis tentang listrik" dan akan merekomendasikan visualisasi yang relevan.

3. **Berbasis Bahasa Alami**: Sistem ini mempertimbangkan konteks input pengguna dan status sistem dalam siklus eksplorasi data, bukan hanya satu kali input kata kunci. Contohnya adalah Eviza dan Evizeon, yang memungkinkan pengguna berinteraksi menggunakan bahasa alami untuk analisis visual.

###**Pendekatan Efisien untuk Visualisasi Data**

Tiga pendekatan utama yang dijelaskan adalah:

1. **Visualisasi Data yang Akurat**: Sistem visualisasi data sering menggunakan SQL untuk mengolah data dari database. Salah satu cara yang umum adalah menerjemahkan permintaan visualisasi menjadi kueri SQL. Namun, ada tantangan dalam mengintegrasikan sistem visualisasi dengan DBMS, seperti duplikasi fungsi dan kesulitan dalam memilih teknik optimasi yang tepat.
Beberapa solusi yang diusulkan termasuk sistem Ermac, yang mengintegrasikan pengambilan dan rendering data untuk mempercepat proses visualisasi.

2. **Penyimpanan Kolom dan Indeks**: Penyimpanan berbasis kolom dapat meningkatkan kinerja untuk aplikasi OLAP, karena pengguna biasanya hanya tertarik pada beberapa kolom data. Indeks juga berperan penting dalam meningkatkan kinerja pencarian dengan mengurangi jumlah data yang perlu diperiksa.
Beberapa teknik yang digunakan termasuk FlashView, yang membangun indeks berbasis pohon hierarkis, dan Falcon, yang menggunakan teknik pengindeksan untuk mengurangi waktu interaksi dalam visualisasi.

3. **Prediksi dan Prefetching**: Dalam eksplorasi data, prediksi data yang mungkin diminati pengguna dan prefetching (pengambilan data sebelumnya) dapat mempercepat proses eksplorasi visualisasi. Teknik ini dibagi menjadi dua kategori: berdasarkan visualisasi yang sedang dieksplorasi dan berdasarkan data historis.
Contohnya termasuk XmdvTool yang memprediksi arah navigasi pengguna berdasarkan riwayat eksplorasi sebelumnya.
Pendekatan-pendekatan ini membantu meningkatkan efisiensi dan kecepatan dalam menciptakan visualisasi data, terutama ketika berurusan dengan dataset besar atau kompleks.

###**Visualisasi Data Pendekatan Aproksimasi**

Ketika volume data meningkat secara eksponensial, modul pemrosesan data tradisional tidak dapat memberikan hasil interaktif yang cepat. Untuk mengatasi masalah ini, digunakan pendekatan Aproksimasi Query Processing (AQP) yang menyediakan hasil visualisasi aproksimasi secara cepat.

* **Pendekatan Berbasis AQP**: Teknik ini menggunakan subset data yang representatif untuk menghasilkan visualisasi dengan cepat, meskipun dengan mengorbankan sedikit kualitas. Contohnya adalah sistem Sample+Seek yang menjawab kueri agregasi dengan kecepatan interaktif dan dalam batas kesalahan yang ditentukan oleh pengguna. Pangloss, sistem berbasis web, menggabungkan visualisasi aproksimasi dengan visualisasi akurat, memungkinkan pengguna mendapatkan wawasan awal dari hasil aproksimasi dan kemudian memverifikasi dengan hasil akurat.

* **Pendekatan Berbasis Sampling Inkremental**: Sistem ini secara bertahap meningkatkan ukuran sampel untuk terus meningkatkan kualitas visualisasi. Contohnya, SampleAction memberikan hasil agregasi parsial dengan batas kesalahan, yang kemudian disempurnakan seiring waktu. IncVisAge memperbaiki masalah fluktuasi signifikan pada visualisasi aproksimasi dengan stabilitas yang lebih baik.

* **Pendekatan Berbasis Persepsi Manusia:** Pendekatan ini berhenti melakukan sampling ketika tidak ada perbedaan yang terlihat secara persepsi antara visualisasi aproksimasi saat ini dan yang mungkin dihasilkan dari sampling lebih lanjut. Contohnya, IFocus menghasilkan bagan batang aproksimasi dengan cepat, memastikan urutan pasangan batang sesuai persepsi manusia. PFunk-H menggabungkan teknik AQP dengan batasan persepsi manusia untuk menghasilkan visualisasi dengan kesalahan yang tidak terdeteksi secara perseptual.

###**Visualisasi Data Progresif**

Visualisasi data progresif memungkinkan hasil visualisasi diperbarui secara bertahap seiring dengan peningkatan data yang diproses. Teknik ini sering digunakan dalam visualisasi aproksimasi dan melibatkan agregasi hierarkis, di mana data dikelompokkan pada berbagai tingkat untuk mendukung eksplorasi visualisasi yang progresif.

1. **Binning Berbasis Rentang**: Sistem seperti imMens memberikan visualisasi dengan resolusi berbeda dengan mengubah ukuran bin. Data multidimensi dibagi menjadi kubus data, yang kemudian dipartisi menjadi tile dengan berbagai tingkat resolusi. Namun, imMens memiliki keterbatasan, seperti tidak memungkinkan pengguna untuk mengubah ukuran bin atau jumlah resolusi.

2. **Binning Berbasis Rentang dan Konten**: HETree-R (berbasis rentang) dan HETree-C (berbasis konten) adalah dua struktur pohon yang digunakan untuk eksplorasi hierarkis. HETree-C, misalnya, memastikan setiap node daun memiliki jumlah data yang sama, cocok untuk skenario seperti klasifikasi nilai ujian. Algoritma konstruksi pohon ini bersifat adaptif dan menyesuaikan dengan preferensi pengguna, mendukung pengalaman eksplorasi yang lebih baik.

###**Spesifikasi Tidak Lengkap dalam Rekomendasi Visualisasi**

Sistem rekomendasi visualisasi dapat bekerja dengan spesifikasi yang kosong (tanpa input pengguna) atau spesifikasi parsial (dengan input parsial dari pengguna). Misalnya, Voyager meminta pengguna memilih satu kolom sebelum merekomendasikan visualisasi, sementara DeepEye menerima input kata kunci seperti "tampilkan grafik garis tentang listrik."

Perbedaan utama antara spesifikasi kosong dan parsial adalah bahwa spesifikasi parsial mempersempit ruang pencarian berdasarkan batasan yang diberikan oleh pengguna. Misalnya, jika pengguna menentukan kolom tertentu, visualisasi yang tidak melibatkan kolom tersebut akan disaring dari kandidat visualisasi.

Dua metode umum yang digunakan untuk memberi peringkat kandidat visualisasi adalah:

1. Peringkat Visualisasi Berbasis Aturan: Sistem ini menggunakan aturan yang telah ditentukan untuk memberi peringkat visualisasi, berdasarkan efektivitas perseptual manusia. Misalnya, Voyager dan DeepEye menggunakan aturan perseptual yang lebih rinci untuk menilai kualitas visualisasi.

2. Peringkat Visualisasi Berbasis Pembelajaran Mesin: Dengan kemajuan pembelajaran mesin, sistem seperti Draco dan DeepEye menggunakan data pelatihan untuk membangun model yang dapat menentukan visualisasi mana yang lebih baik. Draco, misalnya, menggunakan kendala lunak yang dipelajari dari preferensi manusia, sementara DeepEye mengembangkan solusi berbasis pembelajaran mesin untuk menangkap pengetahuan desain visualisasi secara otomatis dari contoh-contoh.

###**Spesifikasi Berbasis Referensi dalam Rekomendasi Visualisasi**

Beberapa sistem rekomendasi visualisasi memberikan rekomendasi berdasarkan data atau visualisasi referensi. Sistem ini biasanya merekomendasikan visualisasi yang mirip atau berbeda dari referensi yang diberikan dalam aspek tertentu.

1. Berbasis Deviasi: SeeDB merekomendasikan visualisasi dengan menghitung deviasi dari visualisasi referensi. Pengguna perlu menentukan kueri dan dataset referensi. SeeDB kemudian menghitung kombinasi variabel, transformasi, tipe mark, dan tipe encoding pada kedua dataset tersebut. Visualisasi dengan deviasi terbesar dari referensi akan direkomendasikan.

2. Berbasis Anomali: Profiler merekomendasikan visualisasi yang paling baik dalam membedakan anomali pada visualisasi utama. Data dalam visualisasi utama diklasifikasikan menjadi poin normal dan abnormal, dan Profiler merekomendasikan visualisasi yang meminimalkan jarak antara klasifikasi dan anomali.

###**Rekomendasi Berbasis Perilaku**

Sistem rekomendasi berbasis perilaku menangkap perilaku pengguna saat ini sebagai input, kemudian menyimpulkan tugas yang diinginkan pengguna dan merekomendasikan visualisasi yang berguna berdasarkan tugas tersebut.

**HARVEST** adalah sistem rekomendasi visualisasi berbasis perilaku yang merekomendasikan visualisasi berdasarkan tugas pengguna yang disimpulkan dari perilakunya. Karena sulit bagi pengguna untuk menjelaskan niat mereka dengan jelas, dan tugas pengguna dapat berkembang selama proses eksplorasi, HARVEST menebak niat pengguna berdasarkan tindakan mereka. HARVEST menangkap tindakan pengguna seperti inspeksi, penyaringan, dan penandaan, yang kemudian dianalisis untuk menentukan konteks tugas dan pola pengguna. HARVEST mengenali empat pola: scan, flip, swap, dan drill-down. Misalnya, jika pengguna secara iteratif memeriksa harga hotel di berbagai wilayah pada peta, HARVEST dapat mendeteksi pola scan, yang menunjukkan bahwa pengguna ingin membandingkan beberapa atribut antara beberapa wilayah.

###**Rekomendasi yang Dipersonalisasi**

Sistem rekomendasi yang dipersonalisasi menangkap perilaku historis pengguna untuk merekomendasikan visualisasi yang sesuai dengan minat mereka.

1. Model Linier: VizDeck adalah sistem rekomendasi visualisasi yang dipersonalisasi yang melatih model linier untuk setiap pengguna berdasarkan perilaku historis mereka. VizDeck menampilkan visualisasi yang direkomendasikan dalam bentuk grid yang disebut "vizlets." Pengguna dapat menjelajah, menata ulang, atau memilih vizlets, dan preferensi visualisasi mereka akan digunakan untuk melatih model yang memberikan rekomendasi di masa depan.

2. Collaborative Filtering: Selain melatih model untuk setiap pengguna, VizRec menggunakan teknik Collaborative Filtering (CF) untuk merekomendasikan visualisasi. VizRec membuat matriks yang mencatat rating pengguna terhadap visualisasi di masa lalu. Berdasarkan kesamaan preferensi dengan pengguna lain, sistem dapat merekomendasikan visualisasi yang mungkin menarik bagi pengguna.

3. Content-based Filtering: Untuk pengguna baru yang belum memiliki data historis, VizRec menggunakan metode rekomendasi berbasis konten. Sistem ini menganalisis fitur-fitur visualisasi dan profil pengguna, kemudian menggunakan frekuensi fitur tersebut untuk membangun profil pengguna dan visualisasi. Kesamaan antara profil pengguna dan kandidat visualisasi kemudian digunakan sebagai skor peringkat.

4. Hybrid Filtering: Metode hibrida menggabungkan kedua pendekatan di atas. VizRec menggunakan jumlah berbobot dari skor normalisasi kedua metode untuk memberikan rekomendasi yang lebih adaptif, terutama jika minat pengguna berubah.

###**Persiapan Data untuk Visualisasi Data**

Data kehidupan nyata sering kali kotor dan visualisasi data yang kotor dapat menyesatkan pengguna. Masalah ini telah lama dikenal sebagai salah satu bentuk visualisasi yang bias. Misalnya, dataset yang diintegrasikan dari berbagai sumber mungkin mengandung duplikat. Oleh karena itu, data yang akan divisualisasikan harus dibersihkan, seperti normalisasi nilai, deduplikasi, imputasi nilai yang hilang, dan deteksi outlier. Tableau telah mengintegrasikan Trifacta untuk persiapan data pada seluruh dataset.

Beberapa studi terkait telah dilakukan oleh komunitas database dan visualisasi untuk menyelidiki dampak data kotor pada visualisasi data:

* Analisis What-if untuk Outlier: Scorpion memungkinkan pengguna secara manual mengidentifikasi outlier dari hasil kueri agregasi. Sistem ini kemudian mencari dan menghilangkan predikat yang menyebabkan outlier tanpa mempengaruhi data lainnya.

* Evaluasi Visualisasi dengan Data yang Hilang: Sebuah studi crowdsourcing mengukur faktor-faktor yang mempengaruhi akurasi respons, kualitas data, dan kepercayaan dalam interpretasi untuk data deret waktu dengan nilai yang hilang. Studi ini mengevaluasi berbagai metode imputasi dan cara menampilkan nilai imputasi untuk melihat pengaruhnya terhadap kualitas dan kepercayaan data.

* Deteksi Visualisasi Bias: Visualisasi yang tampak baik mungkin sebenarnya bias. Oleh karena itu, perlu adanya deteksi otomatis terhadap visualisasi semacam ini, tidak hanya dari perspektif statistik tetapi juga dari sudut pandang data yang kotor.

* Pembersihan Data yang Sesuai dengan Tugas: Membersihkan dataset menjadi lebih mudah jika tugas yang ditargetkan sudah diketahui. Hal ini memungkinkan pembersihan hanya pada sebagian kecil data yang relevan, yang lebih efisien dibandingkan membersihkan seluruh dataset.

###**Tolok Ukur Visualisasi Data**

Seperti halnya ImageNet atau tolok ukur TPC klasik, penting untuk mengembangkan tolok ukur untuk kinerja dan rekomendasi visualisasi data. Tolok ukur ini harus setia pada tugas analisis visual, menyediakan jejak dan data yang dapat digunakan kembali, dan dalam hal rekomendasi, memiliki cakupan dan kualitas label yang tinggi. Ada fokus yang sedang berkembang dalam mengembangkan tolok ukur untuk pengukuran kinerja.

Sebuah penelitian, VizNet, telah memperkenalkan korpus besar berisi lebih dari 31 juta dataset yang dikompilasi dari repositori data terbuka dan galeri visualisasi online. Ini menyediakan dasar yang umum untuk membandingkan teknik desain visualisasi, serta mengembangkan model dan algoritma tolok ukur untuk otomatisasi analisis visual. Namun, masih banyak yang perlu dilakukan.


**Peluang Penelitian**

Kategorisasi visualisasi: Untuk ImageNet, kategori seperti balon atau stroberi mudah ditetapkan karena tugas klasifikasi lebih jelas. Namun, sulit untuk mendefinisikan kategori serupa untuk visualisasi pada tingkat konseptual, seperti tren atau distribusi.
Data pelatihan: Asumsikan kategori bisa ditetapkan, maka tugas yang menantang adalah memberi label pada visualisasi, di mana setiap visualisasi mungkin memiliki banyak label. Setelah itu, masalah sulit lainnya adalah bagaimana menggunakan data berlabel ini, misalnya, dengan menggunakan model pembelajaran mesin atau pembelajaran mendalam mana yang dapat memprediksi visualisasi yang baik untuk tugas tertentu.

###**Visualisasi Data untuk Aplikasi yang Terkait dengan Database**

Seperti yang telah disebutkan di bagian 1, visualisasi data memainkan peran penting dalam aplikasi yang terkait dengan database, seperti Excel, Google Sheets, Oracle Data Visualization Desktop, IBM Db2, Amazon Quicksight, Microsoft Power BI, dan banyak lainnya. Seiring dengan perkembangan teknik visualisasi yang pesat, terdapat lebih banyak peluang untuk menggunakan visualisasi data dalam aplikasi yang terkait dengan database.

**Peluang Penelitian**

1. **Visualisasi data untuk penemuan data**: Penemuan data, yaitu masalah menemukan dataset yang menarik untuk suatu aplikasi tertentu dari data lake yang berisi ribuan atau jutaan data silo, masih menjadi masalah yang sulit untuk dipecahkan. Salah satu hambatannya adalah bagaimana dengan cepat memahami dataset yang ditemukan. Secara praktis, menjelajahi setiap dataset memakan waktu. Secara intuitif, visualisasi data yang memberikan pemahaman tingkat tinggi dapat membantu dalam menyelesaikan masalah penting ini.
